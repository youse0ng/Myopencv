# GMM(Gausian Mixture Model) 가우시안 혼합 모델

- 가우시안 모델 하나만 써서 생성하는 이미지는 노이지를 많이 가지고 있는 채로 나오고, 풍부한, 다양한 부류의 이미지 생성이 불가능하다.
- 사람 얼굴을 생성하는 경우 성별, 인종, 머리 모양과 수염, 안경, 장신구 등의 변화를 수용할 수 있어야한다.
- 단 하나의 가우시안 모델만을 가지고는 불가능하므로, 여러개의 혼합된 가우시안의 어떤 모델을 만들면, 1개 쓰던 것보다 괜찮은 성능을 야기할 수 있지 않을까?
- GMM에 대해서 알아보자

```python
import numpy as np
from tensorflow.keras.datasets import mnist
from sklearn.mixture import GaussianMixture

(x_train,y_train),(x_test,y_test) = mnist.load_data()
X=x_train[np.isin(y_train,[8])]
X = X.reshape(X.shape[0],28*28)
k = 8
gm = GaussianMixture(n_components=k).fit(X)
gen = gm.sample(n_samples=10)

print('WELL DONE So Far')
import matplotlib.pyplot as plt

plt.figure()
for i in range(k):
    plt.subplot(1,10,i+1)
    plt.imshow(gm.means_[i].reshape((28,28)),cmap='gray')
    plt.xticks([])
    plt.yticks([])
plt.show()
```

확실히 하나 가우시안 모델쓰는것 보단, 여러개를 써서 생성형 확률모델을 만드는게 더 효과적인 것 같다.

그렇지만, 한계점은 실제 데이터의 분포가 정규분포와 같은 형상을 띄지 않는다는 점이다. 따라서 그 둘 간의 차이가 커 생성 모델로써는 한계가 있다.
